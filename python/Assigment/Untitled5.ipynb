```python
# Assignment number 7a
# To find the area of different shapes

def case1():
    pi = 3.14
    r = int(input("Enter the radius of circle: "))
    area1 = pi * r * r
    print("Area of circle is :", area1)

def case2():
    l = int(input("Enter the length of rectangle: "))
    b = int(input("Enter the breadth of rectangle: "))
    area2 = l * b
    print("Area of rectangle is :", area2)

def case3():
    a = int(input("Enter the side of square: "))
    area3 = a * a
    print("Area of square is :", area3)

def case4():
    h = int(input("Enter the height of triangle: "))
    c = int(input("Enter the base of triangle: "))
    area4 = 0.5 * h * c
    print("Area of triangle is :", area4)

def case5():
    base = int(input("Enter the base of parallelogram: "))
    height = int(input("Enter the height of parallelogram: "))
    area5 = base * height
    print("Area of parallelogram is :", area5)

switch = {
    1: case1,
    2: case2,
    3: case3,
    4: case4,
    5: case5
}

while True:
    while True:
        print("1.circle", "\n", "2.rectangle", "\n", "3.square", "\n", "4.Triangle", "\n", "5.pallalogram")
        case_number = int(input("Enter case number (1-5), or 0 to exit: "))
        if case_number == 0:
            break
        elif case_number in switch:
            switch[case_number]()
            break
        else:
            print("Invalid case number. Please enter a number between 1 and 5.")
    
    choice = int(input("Do you want to continue (1/0)? "))
    if choice == 0:
        print("Thank you!!")
        break

# Assignment number 7b
class Rectangle:
    def __init__(self, breadth, length):
        self.breadth = breadth
        self.length = length

    def area(self):
        return self.breadth * self.length

a = int(input("Enter length of rectangle: "))
b = int(input("Enter breadth of rectangle: "))
obj = Rectangle(a, b)
print("Area of rectangle:", obj.area())
```
